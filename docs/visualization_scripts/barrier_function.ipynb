```python
import matplotlib.pyplot as plt
import numpy as np

# parameters
alpha = 0.5
constraint = 2

x_0 = -0.1
x_1 = constraint*2

# Function to be modelled
def fx_1(x, alpha, constraint):
    return 1/( -(x)**alpha + constraint**alpha)

def fx_2(x, constraint):
    return -np.log(-x**2 + constraint**2)

def fx_3(x, constraint):
    return np.e**x 

x_range = np.arange(x_0, x_1, 0.01)

y_1, y_2, y_3 = [], [], []
for x in x_range:
    y_1.append(fx_1(x, alpha, constraint))
    y_2.append(fx_2(x, constraint))
    y_3.append(fx_3(x, constraint))

plt.title("Barrier function")
plt.grid(True)
plt.ylabel("f(x)")
plt.xlabel("x")
plt.axis((x_0, x_1, -25, 100))
plt.plot(x_range, y_1, "r")
plt.plot(x_range, y_2, "g")
plt.show()
```
